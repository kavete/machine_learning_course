## Introduction to Matplotlib

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3,4])
```

```python
plt.show()
```

```python
x = [2,2,3,2]
y= [11, 22,33,44]
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure() # creates a figure
ax = fig.add_subplot() #adds axes
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,[50,100,200,500]);
```

```python
plt.show()
```

```python
type(fig), type(ax)
```

```python
x = [1,2,3,4]
y = [11, 22, 33, 44]
fig, ax =plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")
fig.savefig("images/sample-plot.png")
```

```python
plt.show()
```
