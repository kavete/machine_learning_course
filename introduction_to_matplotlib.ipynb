```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3,4])
```

```python
plt.show()
```

```python
x = [2,2,3,2]
y= [11, 22,33,44]
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure() # creates a figure
ax = fig.add_subplot() #adds axes
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,[50,100,200,500]);
```

```python
plt.show()
```

```python
type(fig), type(ax)
```

```python
x = [1,2,3,4]
y = [11, 22, 33, 44]
fig, ax =plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")
fig.savefig("images/sample-plot.png")
```

```python
plt.show()
```

## Figures with NumPy arrays

```python
x = np.linspace(0,10,100)
x [:10]
```

```python
x[:5]
```

```python
fig, ax =plt.subplots()
ax.plot(x, x**2) # line plot
plt.show()
```

```python
#Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
plt.show()
```

```python
fig, ax =plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
plt.show()
```

```python
# Plot from dictionary
nut_butter_prices = { "Almond butter" : 10,
                     "Peanut_butter" : 8,
                     "Cashew butter" : 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter store", ylabel= "Prices ($)")
```
